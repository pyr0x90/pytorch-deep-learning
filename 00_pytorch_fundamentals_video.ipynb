```python
import torch
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Tensors

#### Creating Tensors

```python
#Scalar
scalar = torch.tensor(7)
print(scalar)
print(scalar.ndim)
print(scalar.item())
```

```python
#Vector
vector = torch.tensor((7,7))
print(vector)
print(vector.ndim)
print(vector.shape)
```

```python
#Matrix
MATRIX = torch.tensor([[7,8],
                       [1,2]])
print(MATRIX)
print(MATRIX.ndim)
print(MATRIX.shape)
print(MATRIX.size())
print(MATRIX[0])
print(MATRIX[0][0])
```

```python
#Tensor
TENSOR = torch.tensor([[[1,2,3],
                        [3,4,5],
                        [6,7,8]]])
print(TENSOR)
print(TENSOR.ndim)
print(TENSOR.shape)
print(TENSOR[0])
print(TENSOR[0][0])
print(TENSOR[0][0][0])
```

#### Random Tensors

#### Getting information from tensors

```python
some_tensor = torch.rand(3,4)

print(f"Shape of tensor: {some_tensor.shape}")
print(f"Datatype of tensor: {some_tensor.dtype}")
print(f"Device tensor is stored on: {some_tensor.device}")
```

#### Manupulating Tensors

```python
tensor = torch.tensor([1,2,3])
print(tensor + 10)
print(tensor.add(10))
print(tensor*10)
print(tensor.mul(10))
print(tensor-10)
print(tensor.sub(10))
print(tensor/10)
print(tensor.div(10))
print(tensor.dot(tensor))
print(tensor.dot(tensor.T))
```

#### Matrix Multiplication

```python
# ELement-wise multiplication
tensor = torch.tensor([1,2,3])
tensor*tensor
```

```python
# Matrix multiplication
print(tensor@tensor) #Not recommended
torch.matmul(tensor,tensor)
```

```python
# Shape Errors
tensor_A = torch.tensor([[1,2],
                        [3,4],
                        [5,6]])
tensor_B = torch.tensor([[7,10],
                        [8,11],
                        [9,12]])

print(torch.matmul(tensor_A,tensor_B.T))
print(torch.mm(tensor_A,tensor_B.T))
#torch.matmul(tensor_A,tensor_B)#Error
```

```python
#Inner dimensions must match and the output shape is determined by the outer dimensions
torch.matmul(torch.rand(3,2), torch.rand(2,6))
```

#### Tensor Aggregation

```python
X = torch.arange(0,100,10,dtype = torch.float32)
print(X)
print(X.min())
print(X.max())
print(X.mean())
print(X.argmax())
print(X[X.argmax()])
print(X.argmin())
print(X.std())
```

#### Stack and Reshape Squeeze, Permute

```python
#Create a random tensor of size 3x7x2
x = torch.arange(1.0,10.0,1.0)
print(x)
print(x.shape)
x = x.reshape(3,3)
print(x)
```

```python
#View is a method that returns a new tensor with the same data as the original tensor
z = x.view(1,9)
print(z) #Changing z will change x because they are the same tensor since it is a view
z[0,3] = 100
print(z)
print(x)
```

```python
#Stacking Tensors
print(f'original tensor shape {x.shape}')
x = x.reshape(1,9).squeeze()
print(f'reshaped (1,9) tensor shape {x.shape}')
print(x)
x_stacked = torch.stack([x,x,x],dim = 1) #dim = 0 is vertical growth, dim = 1 is horizontal growth
print('x stacked 3 times, dim = 1:')
print(x_stacked)
print(x_stacked.shape)
```

```python
x = torch.rand(2,3,4)
print(x)
print(f'x shape: {x.shape}')

torch.permute(x, (1,2,0))
```

```python
#Image Tensor
x_original = torch.rand(size=(244,244,3)) #[height, width, channels]
print(x_original.shape)
x_plottable = x_original.permute(2,0,1) #[channels, height, width]
print(x_plottable.shape)
```
