```python
import torch
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Tensors

#### Creating Tensors

```python
#Scalar
scalar = torch.tensor(7)
print(scalar)
print(scalar.ndim)
print(scalar.item())
```

```python
#Vector
vector = torch.tensor((7,7))
print(vector)
print(vector.ndim)
print(vector.shape)
```

```python
#Matrix
MATRIX = torch.tensor([[7,8],
                       [1,2]])
print(MATRIX)
print(MATRIX.ndim)
print(MATRIX.shape)
print(MATRIX.size())
print(MATRIX[0])
print(MATRIX[0][0])
```

```python
#Tensor
TENSOR = torch.tensor([[[1,2,3],
                        [3,4,5],
                        [6,7,8]]])
print(TENSOR)
print(TENSOR.ndim)
print(TENSOR.shape)
print(TENSOR[0])
print(TENSOR[0][0])
print(TENSOR[0][0][0])
```

#### Random Tensors

```python
## Random Tensors
random_tensor = torch.rand(3,244,244) #Random Image Tensor (R,G,B)
#print(random_tensor)
print(random_tensor.shape)
permute_random_tensor = random_tensor.permute(1,2,0)
print(permute_random_tensor.shape)
```

```python
plt.imshow(permute_random_tensor)
```

```python
TENSOR_red = torch.zeros(1,4,4)
TENSOR_blue = torch.zeros(1,4,4)
TENSOR_green = torch.ones(1,4,4)
#TENSOR_img = torch.tensor()
#print(TENSOR)
#print(TENSOR[0:-1])
#print(TENSOR)
print(TENSOR_green)
img = torch.cat((TENSOR_red,TENSOR_green,TENSOR_blue))
```

```python
#print(TENSOR)
plt.imshow(img.permute(1,2,0))
```

#### Zeros and Ones

```python
zeros = torch.zeros((3,4))
randTens = torch.rand((3,4))
print(randTens)
print(zeros*randTens)
```

#### Creating a range and tensors like

```python
rangeVal = torch.arange(0,10,2)
print(rangeVal)
```

```python
tensorLike = torch.zeros_like(rangeVal)
print(tensorLike)
```

#### Tensor Dataypes

```python
float_32_tens = torch.tensor([3.0,6.0,9.0], 
                             dtype = torch.float16, #What dataype is the tensor (e.g. flaot32, float16, float64, double,)
                             device = None, #What device is the tensor on (e.g. cpu, gpu)
                             requires_grad = False)  #Does the tensor require a gradient, which is used in backpropagation

float_32_tens.dtype
```

#### Getting information from tensors

```python
some_tensor = torch.rand(3,4)

print(f"Shape of tensor: {some_tensor.shape}")
print(f"Datatype of tensor: {some_tensor.dtype}")
print(f"Device tensor is stored on: {some_tensor.device}")
```

#### Manupulating Tensors

```python
tensor = torch.tensor([1,2,3])
print(tensor + 10)
print(tensor.add(10))
print(tensor*10)
print(tensor.mul(10))
print(tensor-10)
print(tensor.sub(10))
print(tensor/10)
print(tensor.div(10))
print(tensor.dot(tensor))
print(tensor.dot(tensor.T))
```

#### Matrix Multiplication

```python
# ELement-wise multiplication
tensor = torch.tensor([1,2,3])
tensor*tensor
```

```python
# Matrix multiplication
print(tensor@tensor) #Not recommended
torch.matmul(tensor,tensor)
```

```python
# Shape Errors
tensor_A = torch.tensor([[1,2],
                        [3,4],
                        [5,6]])
tensor_B = torch.tensor([[7,10],
                        [8,11],
                        [9,12]])

print(torch.matmul(tensor_A,tensor_B.T))
torch.matmul(tensor_A,tensor_B)#Error
```

```python
#Inner dimensions must match and the output shape is determined by the outer dimensions
torch.matmul(torch.rand(3,2), torch.rand(2,6))
```
